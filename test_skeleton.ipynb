```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
```

```python
data_features = np.load("./MARS/feature/featuremap_train.npy").reshape(-1, 64, 5)
data_labels = np.load("./MARS/feature/labels_train.npy")
data_labels = np.reshape(data_labels, (-1, 19, 3), order='F')
print(data_features.shape, data_labels.shape)

for idx in tqdm(range(len(data_labels))):

    fig, axs = plt.subplots(2, 3, figsize=(15, 10), subplot_kw={'projection': '3d'})

    axs[0,0].set_visible(False)

    axs[0,1].scatter(data_labels[idx,:, 0], data_labels[idx,:,1], data_labels[idx,:,2])  
    axs[0,1].set_xlim(-2, 2)
    axs[0,1].set_ylim(0, 6)
    axs[0,1].set_zlim(-2, 2)
    axs[0,1].set_xlabel('X')
    axs[0,1].set_ylabel('Y')
    axs[0,1].set_zlabel('Z')
    axs[0,1].view_init(10, -100)
    axs[0,1].set_xticks([-2, -1, 0, 1, 2])
    axs[0,1].set_yticks([0, 2, 4, 6])
    axs[0,1].set_zticks([-2, -1, 0, 1, 2])    


    axs[0,2].scatter(data_features[idx,:, 0], data_features[idx,:,1], data_features[idx,:,2])  
    axs[0,2].set_xlim(-2, 2)
    axs[0,2].set_ylim(0, 6)
    axs[0,2].set_zlim(-2, 2)
    axs[0,2].set_xlabel('X')
    axs[0,2].set_ylabel('Y')
    axs[0,2].set_zlabel('Z')
    axs[0,2].view_init(10, -100)
    axs[0,2].set_xticks([-2, -1, 0, 1, 2])
    axs[0,2].set_yticks([0, 2, 4, 6])
    axs[0,2].set_zticks([-2, -1, 0, 1, 2])    

    axs[1,0].scatter(data_features[idx,:, 0], data_features[idx,:,1], data_features[idx,:,2])
    axs[1,0].set_xlim(-2, 2)
    axs[1,0].set_ylim(0, 6)
    axs[1,0].set_zlim(-2, 2)
    axs[1,0].set_xlabel('X')
    axs[1,0].set_ylabel('Y')
    axs[1,0].set_zlabel('Z')
    axs[1,0].view_init(0, -90)
    axs[1,0].set_xticks([-2, -1, 0, 1, 2])
    axs[1,0].set_yticks([0, 2, 4, 6])
    axs[1,0].set_zticks([-2, -1, 0, 1, 2])


    axs[1,1].scatter(data_features[idx,:, 0], data_features[idx,:,1], data_features[idx,:,2])
    axs[1,1].set_xlim(-2, 2)
    axs[1,1].set_ylim(0, 6)
    axs[1,1].set_zlim(-2, 2)
    axs[1,1].set_xlabel('X')
    axs[1,1].set_ylabel('Y')
    axs[1,1].set_zlabel('Z')
    axs[1,1].view_init(90, -90)
    axs[1,1].set_xticks([-2, -1, 0, 1, 2])
    axs[1,1].set_yticks([0, 2, 4, 6])
    axs[1,1].set_zticks([-2, -1, 0, 1, 2])


    axs[1,2].scatter(data_features[idx,:, 0], data_features[idx,:,1], data_features[idx,:,2])
    axs[1,2].set_xlim(-2, 2)
    axs[1,2].set_ylim(0, 6)
    axs[1,2].set_zlim(-2, 2)
    axs[1,2].set_xlabel('X')
    axs[1,2].set_ylabel('Y')
    axs[1,2].set_zlabel('Z')
    axs[1,2].view_init(0, 180)
    axs[1,2].set_xticks([-2, -1, 0, 1, 2])
    axs[1,2].set_yticks([0, 2, 4, 6])
    axs[1,2].set_zticks([-2, -1, 0, 1, 2])

    # plt.show
    plt.savefig("../Self Collected Data/images/frameYZ_{}.jpg".format(idx))
    plt.close(fig)
```
